```python
def get_stores_pages():
    # grab items and put them in a list 
    stores = []
    
    # Set the URL and Endpoint 
    domain = 'https://python.zgulde.net'
    endpoint_store = '/api/v1/stores'
    
    if os.path.isfile('stores.csv') == False:
    
        while endpoint_store != None:
            # Set the url for the loop
            url = domain + endpoint_store


            # set the request to response 
            response = requests.get(url)

            
            #defining data
            data = response.json()     

            # Set current page
            page = data['payload']['page']
            print(f'current page is: {page}')

            # Set the next page
            endpoint_store = (data['payload']['next_page'])

            # Set the last page
            last_page = (data['payload']['max_page'])

            # .extend adds elemnts from a list to another list
            stores.extend(data['payload']['stores'])
    
        # make items into a dataframe
        stores = pd.DataFrame(stores)

            # Create a csv
        stores.to_csv('stores.csv')

    else:
        #If it is already made this just reads the csv
        stores = pd.read_csv('stores.csv', index_col=0)

    return stores
```

```python
def get_sales_pages():
    # grab items and put them in a list 
    sales = []
    
    # Set the URL and Endpoint 
    domain = 'https://python.zgulde.net'
    endpoint_sale = '/api/v1/sales'
    
    if os.path.isfile('sales.csv') == False:
    
        while endpoint_sale != None:
            # Set the url for the loop
            url = domain + endpoint_sale


            # set the request to response 
            response = requests.get(url)

            
            #defining data
            data = response.json()     

            # Set current page
            page = data['payload']['page']
            print(f'current page is: {page}')

            # Set the next page
            endpoint_sale = data['payload']['next_page']

            # Set the last page
            last_page = (data['payload']['max_page'])

            # .extend adds elemnts from a list to another list
            sales.extend(data['payload']['sales'])

        # make items into a dataframe
        sales = pd.DataFrame(sales)

        # Create a csv
        sales.to_csv('sales.csv')

    else:
        #If it is already made this just reads the csv
        sales = pd.read_csv('sales.csv', index_col=0)

    return sales
```
