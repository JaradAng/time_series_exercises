```python

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd


import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime

from sklearn.metrics import mean_squared_error
from math import sqrt 

import statsmodels.api as sm
from statsmodels.tsa.api import Holt
```

```python
# acquire data
df = pd.read_csv('https://ds.codeup.com/saas.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# set train size to be 60% of total 
train_size = int(len(df) * 0.6)
train_size
```

```python
# set validate size to be 30% of total 
validate_size = int(len(df) * 0.15)
validate_size
```

```python
# set test size to be number of rows remaining. 
test_size = int(len(df) - train_size - validate_size)
test_size
```

```python
# validate will go from 24 to 24+14
validate_end_index = train_size + validate_size
validate_end_index
```

```python
# train will go from 0 to 24
train = df[:train_size]
# validate will go from 24 to 38
validate = df[train_size:validate_end_index]
# test will include 38 to the end
test = df[validate_end_index:]
```

```python
for col in train.columns:
    plt.figure(figsize=(14,8))
    plt.plot(train[col])
    plt.plot(validate[col])
    plt.plot(test[col])
    plt.ylabel(col)
    plt.title(col)
    plt.show()
```
